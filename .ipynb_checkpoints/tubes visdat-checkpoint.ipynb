```python
import pandas as pd

# Baca dataset COVID
df = pd.read_csv("covid_19_indonesia_time_series_all.csv")

# Pastikan nama kolom bersih dari spasi dan huruf kecil semua
df.columns = df.columns.str.strip().str.lower()

# Ubah format kolom 'date' ke datetime jika tersedia
if 'date' in df.columns:
    df['date'] = pd.to_datetime(df['date'])
else:
    raise KeyError("Kolom 'date' tidak ditemukan di dataset. Cek kembali nama kolomnya.")

# Lihat 5 baris pertama untuk verifikasi
print(df.head())
```

```python
from bokeh.io import curdoc
from bokeh.models import ColumnDataSource, Select, HoverTool, Div
from bokeh.layouts import column, row
from bokeh.plotting import figure
import pandas as pd
from math import pi

# --- LOAD & CLEAN DATA ---
df = pd.read_csv("covid_19_indonesia_time_series_all.csv")
df.columns = df.columns.str.strip().str.lower()  # pastikan kolom bersih
if 'date' in df.columns:
    df['date'] = pd.to_datetime(df['date'])
else:
    raise KeyError("Kolom 'date' tidak ditemukan!")

# Rename agar sesuai dengan struktur dashboard
df.rename(columns={
    'date': 'Date',
    'location': 'Location',
    'total_cases': 'Total Cases',
    'total_deaths': 'Total Deaths',
    'total_recover': 'Total Recovered'
}, inplace=True)

# --- DROPDOWN OPTIONS ---
provinsi_list = sorted(df['Location'].unique())
default_prov = 'Indonesia' if 'Indonesia' in provinsi_list else provinsi_list[0]

# Data terkini untuk bar chart
df_latest = df[df['Date'] == df['Date'].max()]
df_grouped = df_latest[df_latest['Location'] != 'Indonesia']
df_allprov = df_grouped.groupby('Location')['Total Cases'].max().sort_values(ascending=False).reset_index()

# --- DATA SOURCES ---
source_line = ColumnDataSource(data=dict(x=[], y=[]))
source_pie = ColumnDataSource(data=dict(Kategori=[], Jumlah=[], color=[], start_angle=[], end_angle=[]))
source_bar = ColumnDataSource(data=dict(provinsi=[], jumlah=[]))

# --- LINE CHART ---
fig_line = figure(x_axis_type="datetime", height=300, sizing_mode="stretch_width")
fig_line.line('x', 'y', source=source_line, line_width=2, color="navy", legend_label="Kasus Baru")
fig_line.circle('x', 'y', source=source_line, size=5, color="navy", alpha=0.5)
fig_line.xaxis.axis_label = "Tanggal"
fig_line.yaxis.axis_label = "Jumlah Kasus Baru"
fig_line.add_tools(HoverTool(tooltips=[("Tanggal", "@x{%F}"), ("Kasus Baru", "@y")],
                              formatters={'@x': 'datetime'}))

# --- PIE CHART ---
fig_pie = figure(height=300, title="", toolbar_location=None,
                 tools="hover", tooltips="@Kategori: @Jumlah", x_range=(-0.5, 1.0), sizing_mode="stretch_both")
fig_pie.wedge(x=0, y=1, radius=0.4,
              start_angle='start_angle', end_angle='end_angle',
              line_color="white", fill_color='color', legend_field='Kategori', source=source_pie)

# --- BAR CHART ---
fig_bar = figure(y_range=[], height=300, title="", toolbar_location=None,
                 tools="hover", tooltips=[("Provinsi", "@provinsi"), ("Jumlah", "@jumlah")],
                 sizing_mode="stretch_both")
fig_bar.hbar(y='provinsi', right='jumlah', height=0.5, source=source_bar, color="dodgerblue")
fig_bar.xaxis.axis_label = "Jumlah Kasus"
fig_bar.yaxis.axis_label = "Provinsi"

# --- SELECT DROPDOWN ---
provinsi_select = Select(title="Pilih Provinsi", value=default_prov, options=provinsi_list)

# --- CALLBACK FUNCTION ---
def update_data(attr, old, new):
    prov = provinsi_select.value
    df_sel = df[df['Location'] == prov]

    if df_sel.empty:
        return

    df_sel = df_sel.sort_values("Date")
    new_cases = df_sel['Total Cases'].diff().fillna(0)
    source_line.data = {'x': df_sel['Date'], 'y': new_cases}
    fig_line.title.text = f"📈 Tren Kasus Harian COVID-19 di {prov}"

    # Pie Chart
    last_row = df_sel.iloc[-1]
    recovered = last_row['Total Recovered']
    deaths = last_row['Total Deaths']
    active = last_row['Total Cases'] - recovered - deaths
    values = [recovered, deaths, active]
    labels = ['Sembuh', 'Meninggal', 'Aktif']
    colors = ['#28a745', '#dc3545', '#ffc107']
    angles = [v / sum(values) * 2 * pi for v in values]
    start = [sum(angles[:i]) for i in range(len(angles))]
    end = [sum(angles[:i+1]) for i in range(len(angles))]
    source_pie.data = {
        'Kategori': labels,
        'Jumlah': values,
        'color': colors,
        'start_angle': start,
        'end_angle': end
    }
    fig_pie.title.text = f"Proporsi Kasus Terbaru di {prov}"

    # Bar Chart
    top10 = df_allprov.head(10)
    fig_bar.y_range.factors = list(top10['Location'])
    source_bar.data = {'provinsi': top10['Location'], 'jumlah': top10['Total Cases']}
    fig_bar.title.text = "Total Kasus COVID-19 per Provinsi (Top 10)"

# --- REGISTER CALLBACK ---
provinsi_select.on_change('value', update_data)

# --- INITIAL LOAD ---
update_data(None, None, None)

# --- FINAL LAYOUT ---
layout = column(
    Div(text="<h2 style='text-align:center;'>🦠 Dashboard Interaktif COVID-19 Indonesia</h2>"),
    provinsi_select,
    fig_line,
    row(column(fig_pie, sizing_mode="stretch_both"),
        column(fig_bar, sizing_mode="stretch_both"),
        sizing_mode="stretch_width"),
    sizing_mode="stretch_both"
)

curdoc().add_root(layout)
curdoc().title = "Dashboard COVID-19 Indonesia"
```
